```python
traducao = {
    'Wheat & Rye (Bread)': 'Trigo',
    'Maize (Meal)': '',
    'Barley (Beer)': 'Cevada',
    'Oatmeal': 'Aveia branca',
    'Rice': 'Arroz',
    'Potatoes': 'Batata',
    'Cassava': '',
    'Cane Sugar': 'Cana-de-açúcar',
    'Beet Sugar': '',
    'Other Pulses': '',
    'Peas': 'Ervilha em grão',
    'Nuts': 'Noz',
    'Groundnuts': 'Amendoim',
    'Soymilk': '',
    'Tofu': '',
    'Soybean Oil': '',
    'Palm Oil': '',
    'Sunflower Oil': '',
    'Rapeseed Oil': '',
    'Olive Oil': '',
    'Tomatoes': 'Tomate',
    'Onions & Leeks': 'Cebola',
    'Root Vegetables': 'Mandioca',
    'Brassicas': '',
    'Other Vegetables': '',
    'Citrus Fruit': 'Laranja',
    'Bananas': 'Banana',
    'Apples': 'Maçã',
    'Berries & Grapes': 'Uva',
    'Wine': '',
    'Other Fruit': '',
    'Coffee': 'Café',
    'Dark Chocolate': '',
    'Beef (beef herd)': 'Bovinos (Bois e Vacas)',
    'Lamb & Mutton': 'Ovinos (Ovelhas e Carneiros)',
    'Pig Meat': 'Suínos (Porcos)',
    'Poultry Meat': '',
    'Milk': '',
    'Cheese': '',
    'Eggs': 'Galináceos',
    'Fish (farmed)': '',
    'Shrimps (farmed)': ''
}
```

```python
df_food_prod = df_food_production.copy()

df_food_prod["Produto"] = df_food_production['Food product']

df_food_prod = df_food_prod.replace({'Produto': traducao})

df_food_prod.drop(df_food_prod[df_food_prod['Produto'] == ''].index, inplace=True)

df_food_prod = df_food_prod.sort_values(by='Total_emissions', ascending=False)

df_food_prod
```

```python
df_food_prod = df_food_prod.rename(columns={"Produto": "produto"})

dfs_join_agricultura = pd.merge(df_food_prod, df_ranking_agricultura, how="left", on=["produto"])

dfs_join_agricultura.drop(dfs_join_agricultura[dfs_join_agricultura['valor'].isnull()].index, inplace=True)

dfs_join_agricultura
```

```python
dfs_join_pecuaria = pd.merge(df_food_prod, df_ranking_pecuaria, how="left", on=["produto"])

dfs_join_pecuaria.drop(dfs_join_pecuaria[dfs_join_pecuaria['valor'].isnull()].index, inplace=True)

dfs_join_pecuaria
```

```python
teste = dfs_join_pecuaria[['produto','Total_emissions']].pivot(columns='produto', values='Total_emissions')
```

```python
fig, ax = plt.subplots(figsize =(12, 5))
 
# Horizontal Bar Plot
plt.bar(dfs_join_pecuaria['produto'], dfs_join_pecuaria['Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)'])
 
ax.set_title('Emissão de gases do efeito estufa na pecuária por produto', loc ='left', )
plt.ylabel('KgCO2 por 1000kcal')
    
# Show Plot
plt.show()
```

```python
dfs_join_agricultura = dfs_join_agricultura.sort_values(by='Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)', ascending=False)
```
